```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
data = pd.read_csv('/content/Population.csv')
```

```python
# Display the first few rows to understand the structure
data.head()
```

```python
# Filter data for the selected country
country_name = "India"  # Replace with your country of interest
country_data = data[data['Country Name'] == country_name]

# Transpose data to make years the index
country_trend = country_data.loc[:, '1960':'2023'].transpose()
country_trend.columns = ['Population']
country_trend.index.name = 'Year'
country_trend.reset_index(inplace=True)

# Display the transposed data
country_trend.head()
```

```python
# Plot the population trend
plt.figure(figsize=(10, 6))
plt.plot(country_trend['Year'], country_trend['Population'], marker='o', linestyle='-', color='blue')

# Add titles and labels
plt.title(f'Population Trend for {country_name} (1960-2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.grid(alpha=0.3)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Plot the population trend as a bar chart
plt.figure(figsize=(12, 6))
plt.bar(country_trend['Year'], country_trend['Population'], color='skyblue')

# Add titles and labels
plt.title(f'Population Trend for {country_name} (1960-2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45, fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
